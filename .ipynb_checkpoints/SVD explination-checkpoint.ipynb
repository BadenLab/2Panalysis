## Example of SVD usage
I quickly wrote this notebook shortly after recieving some help from Tom. It summarises what he explained to me and how to interpret the resulting   

```python

import numpy as np 
import matplotlib.pyplot as plt
```

```python
def gaus(x, a, b, c):
        # a Gaussian distribution
            return a * np.exp(-(x-b)**2/(2*c**2))
```

```python
## a is a gaussian 
a = gaus(np.arange(0, 11), 1, 4, 12)
## We can optionally tile this gaussian (e.g. repeat it x amount if times
# a = numpy.tile(a, 5)
## We can also optionally add noise
# a_noise = np.random.normal(0, 0, len(a))
# a = a + a_noise
## Then we reshape for dot product multiplication
a = a.reshape(1, len(a))
```

```python
## b is a gaussian 
b = gaus(np.arange(0, 10), 1, 1, 1)+1
## We can optionally tile this gaussian (e.g. repeat it x amount if times
b = numpy.tile(b, 5)
## We can also optionally add noise
# b_noise = np.random.normal(0, 1, len(b))
# b = b + b_noise
## Then we reshape for dot product multiplication
b = b.reshape(len(b), 1)
```

```python
%matplotlib widget
## Then we plot our two lovely example data
plt.plot(a.T, c = 'r') # a is always the red trace
plt.plot(b, c = 'b') # b is always the blue trace
```

```python
%matplotlib widget
## Then we get the dot product of a x b
c = b @ a # here b x a because of reshaping above, only transposes matrix so not a big deal
plt.imshow(c)
```

```python
%matplotlib widget
## Then we generate some noise to add on top, otherwise SVD is sort of pointless (may not extract any single value because model is perfect) 
c_noise = np.random.normal(0, 0.11, c.shape)
c_w_noise = c + c_noise
plt.imshow(c_w_noise)
```

```python
## We then run svd algorithm to get the unitary matrix (u), diagonal matrix (s, aka sigma), and the complex unitary matrix (vh, aka v)
u, s, vh = np.linalg.svd(c_w_noise)
```

```python
%matplotlib widget
## u looks like this
plt.imshow(u)
## The "line" in the middle of the matrix happens because the dimmensions are not the same, so essentially those indeces are spurious 
### You can actually see how the plot above "fits into" between 0 - 10 on both axes in the plot below
#### u describes in hierarchical order the singular values of 1 component of our matrix c
```

```python
%matplotlib widget
## sigma looks like this
plt.plot(s)
## sigma describes how much of the variance of c the indeces of u x vh describes 
## (in our case, because the data is simple, the 1st index describes all the data, and the rest of the indeces asympotically approach 0 in explanotry power)
### Kind of the least important one to understand. Just read it as a measure of which indeces have the most explanatory power. Also does some important role in the actual SVD computation:s
```

```python
%matplotlib widget
## and v looks like this
plt.imshow(vh)
## I don't fully understand what it represents, other than it kind of "mixes" values from u to reverse generate X. (you'll see what I mean below)
## Again, this is hierarchically ordered from most-to-least important/explanatory, but the axes are to my understanding arbitrary
### Note that vh is 10x10 (vector a is 10 long) and u is 50x50 (vector b is 50 long). 
```

```python
%matplotlib widget
## The left column of the unitary matrix u describes the first component of the matrix c
plt.plot(u[:, 0] * -6, linestyle = 'dashed')  # Sometimes the sign flips due to multiplication (or something...?) but this is expected. I also scalled it 6x so it does not look flat compared to the original!
plt.plot(b, c = 'b') # Oragne, dotted trace is the original input vector b for our matrix, c
plt.plot(u[0], c = 'g') # The top row is obviously rubbish
```

```python
%matplotlib widget
## The top row of the complex matrix v or vh describes the other component of the matrix
plt.plot(vh[0] * -3, linestyle = 'dashed', ) # Again flip the sign and multiply to equalise the scale between the two
plt.plot(a.T, c='r') # Oragne, dotted trace is the original input vector a for our matrix, c
plt.plot((vh[:, 0]+2)*.25, c = 'g') # The left column is obviously rubbish (scaled down and shifted up a bit for illustration purposes) 
```
