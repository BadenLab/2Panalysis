```python
import Import_Igor
import matplotlib.pyplot as plt
import numpy as np

import ipywidgets
from __future__ import print_function
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
```

```python
stack = Import_Igor.get_stack(r"C:\Users\Simen\Downloads\nat_mono_2Hz_0.smp")
# stack = Import_Igor.get_stack(r"C:\Users\Simen\Downloads\test_resolutions\256.smh")
stack_ch1 = stack.ch_arrays(129)[0]
stack_ch1.shape
```

```python
# Cropping using LUT resolution_map in class 
%matplotlib widget
fig = plt.figure()
def show_frame(frame):
    plt.imshow(stack_ch1[frame])
    fig.canvas.draw()
interact(show_frame, frame= widgets.IntSlider(min=0, max=len(stack_ch1)-1, step=1, value=1, continuous_update=True))
print(stack_ch1.shape)
```
